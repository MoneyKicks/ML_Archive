```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
%matplotlib inline
```

```python
image=cv2.imread('/home/pawan/Desktop/demo.jpg')
row=100
col=int(row*image.shape[1]/image.shape[0])
im=cv2.resize(image,(col,row))
plt.imshow(im)
plt.show()
```

```python
print(np.shape(im))
im=np.reshape(im,(-1,3))
print(np.shape(im))
```

```python
for k in range(1,25):
    km=KMeans(n_clusters=k)
    km.fit(im)
    centroid=np.array(km.cluster_centers_,dtype=np.uint8)
    freq=np.unique(km.labels_,return_counts=True)
    label_id=freq[0][freq[1].argmax()]
    dominant_color=centroid[label_id]
    stamp=np.zeros((100,100,3),dtype=np.uint8)
    stamp[:,:,:]=dominant_color
    plt.imshow(stamp)
    plt.show()
```
